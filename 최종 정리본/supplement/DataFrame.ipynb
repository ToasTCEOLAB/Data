```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
resultList = []
```

```python
temp1 = ["VGG19", 0.6926, 0.5143, 0.6926, 0.5172]
temp2 = ["ResNet50V2", 0.0525, 0.9786, 2.6478, 0.6552]
temp3 = ["InceptionResNetV2", 0.2238, 0.9143, 1.4126, 0.6552]
temp4 = ["ResNet152V2", 0.2465, 0.9143, 2.6114, 0.5862]
temp5 = ["Xception", 0.0867, 0.9857, 2.1419, 0.5862]
temp6 = ["ResNet50", 0.0001, 1.0000, 2.6841, 0.6379]
temp7 = ["MobileNetV2", 0.0989, 0.9714, 3.7973, 0.4828]
temp8 = ["EfficientNetV2L", 0.6059, 0.6286, 5.9764, 0.5172]

resultList.append(temp1)
resultList.append(temp2)
resultList.append(temp3)
resultList.append(temp4)
resultList.append(temp5)
resultList.append(temp6)
resultList.append(temp7)
resultList.append(temp8)
resultList
```

```python
df = pd.DataFrame(resultList)
df.columns = ["Model Name", "Loss", "Accuracy", "val_Loss", "val_Accuracy"]
df
```

```python
ax = sns.barplot(data=df, x="Model Name", y="Loss")
ax.set_title("Loss")
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)
plt.show()

ax = sns.barplot(data=df, x="Model Name", y="Accuracy")
ax.set_title("Accuracy")
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)
plt.show()

ax = sns.barplot(data=df, x="Model Name", y="val_Loss")
ax.set_title("val_Loss")
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)
plt.show()

ax = sns.barplot(data=df, x="Model Name", y="val_Accuracy")
ax.set_title("val_Accuracy")
ax.set_xticklabels(ax.get_xticklabels(), rotation = 45)
plt.show()
```
