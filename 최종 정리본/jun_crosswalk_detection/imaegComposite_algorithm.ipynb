```python
import cv2
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import os
```

# Done


```python
folder = "C:/Users/user/Desktop/result/";
T = "outside/"
S = "augmentation/";
SP = 'crosswalk_detection/'
P = "person_detection/";
C="composite/"
SP_filenames = os.listdir(folder+SP +T)
# P_filenames = os.listdir(folder+P +T)
```

```python
for i in range(len(SP_filenames)):
    temp = SP_filenames[i]
    temp = SP_filenames[i].replace(".png","_person_0.jpg")
    
    print(SP_filenames[i])
    print(temp)
    black = cv2.imread(folder+SP+ T+SP_filenames[i])
    red = cv2.imread(folder+P+ T+ temp)
    
    red = cv2.resize(red, dsize=(1000, 1000), interpolation=cv2.INTER_LINEAR)
    ret ,mask = cv2.threshold(red,100,255,cv2.THRESH_BINARY_INV)
    b,g,r = cv2.split(mask)

    bbg_img = cv2.merge((r,r,b))
    w_img = cv2.merge((g,g,g))


    b = 255-black
    ret ,w_mask = cv2.threshold(cv2.add(b,w_img),150,255,cv2.THRESH_BINARY_INV)

    IMG = cv2.add(w_mask,bbg_img)
    cv2.imwrite(folder+C+T+temp, IMG)
```
